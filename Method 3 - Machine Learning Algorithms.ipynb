## 3.3.1 Preprocessing and Feature Building

```python
import pandas as pd
#read the data in csv
df_sales = pd.read_csv('SalesData.csv')

#convert date field from string to datetime
df_sales['date'] = pd.to_datetime(df_sales['date'])

#show first 10 rows
#df_sales.head(10)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')

# for Month
month_df = pd.DataFrame(enc.fit_transform(df_sales[['Month']]).toarray())
```

```python
df_sales = df_sales.join(month_df)
#df_sales
```

```python
from sklearn.preprocessing import OneHotEncoder

# creating instance of one-hot-encoder
encd = OneHotEncoder(handle_unknown='ignore')

# for day of the week
day_df = pd.DataFrame(encd.fit_transform(df_sales[['Day_C']]).toarray())
```

```python
df_salesmd = pd.merge(df_sales, day_df, left_index=True, right_index=True)
df_salesmd
```

```python
df_salesmd.to_csv('one hot dataset.csv', index = True)
```

## ML Models (Section 3.3.2 to 3.3.7)

```python
from sklearn.linear_model import *
from sklearn.ensemble import *
from sklearn.metrics import *
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#from sklearn.ensemble import VotingRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.neural_network import MLPRegressor
```

```python
# Linear Regression
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X_train, Y_train)
Y_pred=regr.predict(X_test)
regr_accuracy=r2_score(Y_pred,Y_test)

print('Accuracy', regr_accuracy)
mse = ((Y_pred - Y_test) ** 2).mean()
print('The Mean Squared Error is {}'.format(round(mse, 2)))
print('The Root Mean Squared Error is {}'.format(round(np.sqrt(mse), 2)))
```

```python
#DecisionTreeRegressor
model_DTR = DecisionTreeRegressor()
model_DTR.fit(X_train,Y_train)
Y_predict=model_DTR.predict(X_test)
model_DTR_accuracy=r2_score(Y_predict,Y_test)
model_DTR_accuracy

print ('Accuracy', model_DTR_accuracy)
mse = ((Y_predict - Y_test) ** 2).mean()
print('The Mean Squared Error is {}'.format(round(mse, 2)))
print('The Root Mean Squared Error is {}'.format(round(np.sqrt(mse), 2)))
```

```python
#RandomForestRegressor
model_RFR=RandomForestRegressor()
model_RFR.fit(X_train,Y_train)
Y_predict=model_RFR.predict(X_test)
model_RFR_accuracy=r2_score(Y_predict,Y_test)

print('Accuracy', model_RFR_accuracy)
mse = ((Y_predict - Y_test) ** 2).mean()
print('The Mean Squared Error is {}'.format(round(mse, 2)))
print('The Root Mean Squared Error is {}'.format(round(np.sqrt(mse), 2)))
```

```python
#VotingRegressor
from sklearn.ensemble import VotingRegressor
model_VR=VotingRegressor(estimators=[('g',model_gradient),('rf',model_RFR)])
model_VR.fit(X_train,Y_train)
Y_predict=model_VR.predict(X_test)
model_VR_accuracy=r2_score(Y_predict,Y_test)

print('Accuracy', model_VR_accuracy)
mse = ((Y_predict - Y_test) ** 2).mean()
print('The Mean Squared Error is {}'.format(round(mse, 2)))
print('The Root Mean Squared Error is {}'.format(round(np.sqrt(mse), 2)))
```

```python
#Gradient Boosting Regressor
model_gradient=GradientBoostingRegressor()
model_gradient.fit(X_train,Y_train)
Y_predict=model_gradient.predict(X_test)
model_gradient_accuracy=r2_score(Y_predict,Y_test)

print('Accuracy', model_gradient_accuracy)
mse = ((Y_predict - Y_test) ** 2).mean()
print('The Mean Squared Error is {}'.format(round(mse, 2)))
print('The Root Mean Squared Error is {}'.format(round(np.sqrt(mse), 2)))
```
